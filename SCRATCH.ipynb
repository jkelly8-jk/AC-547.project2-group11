```python
import pandas as pd

# --- load files (adjust name as needed) ---
degree = pd.read_csv("degree.csv")
occ_ind = pd.read_csv("occupation_industry.csv")

# --- helper to standardize column names ---
def clean_columns(df):
    df.columns = df.columns.str.lower().str.strip()
    return df

degree = clean_columns(degree)
occ_ind = clean_columns(occ_ind)

# drop duplicates
degree = degree.drop_duplicates()
occ_ind = occ_ind.drop_duplicates()
```

```python
degree = degree[[
    "year", "serial", "pernum", "degfield", "degfieldd",
    "degfield2", "degfield2d", "educ", "educd", "perwt"
]]

occ_ind = occ_ind[[
    "year", "serial", "pernum", "occ", "ind", "perwt"
]]
```

```python
degree = degree.sort_values(["year", "serial", "pernum"])
occ_ind = occ_ind.sort_values(["year", "serial", "pernum"])

degree = degree.set_index(["year", "serial", "pernum"])
occ_ind = occ_ind.set_index(["year", "serial", "pernum"])
```

```python
degree["is_accounting_degree"] = (degree["degfield"] == 62).astype(int)
```

```python
occ_ind["is_accountant_occ"] = occ_ind["occ"].isin([1600, 1610]).astype(int)

```

```python
occ_ind["is_public_accounting_ind"] = occ_ind["ind"].isin([7870, 7880]).astype(int)
```

```python
deg_trend = (
    degree.groupby("year")
          .apply(lambda x: (x["is_accounting_degree"] * x["perwt"]).sum())
          .reset_index(name="accounting_degree_population")
)
```

```python
occ_trend = (
    occ_ind.groupby("year")
           .apply(lambda x: (x["is_accountant_occ"] * x["perwt"]).sum())
           .reset_index(name="accountant_occupation_population")
)
```

```python
ind_trend = (
    occ_ind.groupby("year")
           .apply(lambda x: (x["is_public_accounting_ind"] * x["perwt"]).sum())
           .reset_index(name="public_accounting_population")
)
```

```python
merged = degree.join(occ_ind, how="inner", rsuffix="_occ")
```

```python
import matplotlib.pyplot as plt

# 1. Merge all three trend tables on year
trend_all = (
    deg_trend
    .merge(occ_trend, on="year", how="outer")
    .merge(ind_trend, on="year", how="outer")
    .sort_values("year")
)

# 2. Plot all trends on one line graph
plt.figure(figsize=(10,6))

plt.plot(
    trend_all["year"],
    trend_all["accounting_degree_population"],
    label="Accounting degrees"
)
plt.plot(
    trend_all["year"],
    trend_all["accountant_occupation_population"],
    label="Accounting occupations"
)
plt.plot(
    trend_all["year"],
    trend_all["public_accounting_population"],
    label="Public accounting industry"
)

plt.title("Accounting Degrees vs Accounting Jobs Over Time")
plt.xlabel("Year")
plt.ylabel("Weighted population")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

```
